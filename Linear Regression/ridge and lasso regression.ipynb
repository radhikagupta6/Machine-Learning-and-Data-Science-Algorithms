```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = load_boston()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
print(dataset.head())
```

```python
dataset.columns = df.feature_names
```

```python
dataset['Price']= df.target
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

### Linear regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
mean_squarred_error = cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=5)
# we get five values from 5 fold cross validation, hence we'd take a mean. The more the final value is near zero the better is the result since the value of cost function is reduced. 
mean_mse = np.mean(mean_squarred_error)
print(mean_mse)
```

### Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_reg = GridSearchCV(ridge, parameters,scoring='neg_mean_squared_error', cv=5)
ridge_reg.fit(X,y)
```

```python
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

### Lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_reg = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv=5)
lasso_reg.fit(X,y)
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

Usually Lasso performs well when we have more number of features
